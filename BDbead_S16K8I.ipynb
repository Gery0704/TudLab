```
#Szükséges importok elvégzése

import pandas as pad
import numpy as nup
import matplotlib.pyplot as plt
```

```
#trainingset CSV beolvasása


# File elérése
file_location = "/FileStore/tables/ida_2016_training_set_update.csv"
file_type = "csv"

# Fontos, állítsuk be, hogy az első sor a fejléc
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

training_set = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(training_set)
```

```
#testset CSV beolvasása akárcsak a training set CSV esetében

file_location = "/FileStore/tables/ida_2016_test_set_update.csv"
file_type = "csv"


infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

test_set = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(test_set)
```

```
#nézzük meg az egyik oszlop első öt sorát
test_set.select(['ab_000']).show(5)
```

```
#az adatoszlopokban található NA értékek zavarhatják a munkánkat, így ezeket hagyjuk el, legyen -1 a helyükön, így nem zavarhatják a munkánkat mivel a #többi értékünk pozitiv, ezt a műveletet a replace használatával megtehetjük

training_set=training_set.replace('na', '-1')
test_set=test_set.replace('na', '-1')
```

```
#Kiszedjük az esetleges üres sorokat a dropna segítségével

print(test_set.count())
test_set = test_set.dropna()
print(test_set.count())

print(training_set.count())
training_set = training_set.dropna()
print(training_set.count())
```

```
#A gyorsabb számítások érdekében konvertáljuk az adatokat double típusra, erre azért van szükség mert így a korábbi (Titanic) megoldásunkkal ellentétben #nem szükséges az összes oszlopra indexer illetve encoder használata
#Ez nagyban meggyorsítja a munkánkat hiszen a feladat leírása alapján mi szeretnénk az összes érték alapján előrejelzéseinket előállítani, ami ~170 
#esetében oszloponként egy indexer és egy encoder használatával járna 

from pyspark.sql.types import DoubleType
for e in training_set.columns[1:]:
  training_set = training_set.withColumn(e, training_set[e].cast(DoubleType()))
```

```
#Osszuk el a split sehítségével a két konténerünkbe az adatokat
(traindf,testdf) = training_set.randomSplit([0.8,0.2])
```

```
#importáljuk be a szükséges könyvtárakat

from pyspark.ml.feature import StringIndexer, VectorIndexer, VectorAssembler, OneHotEncoderEstimator,IndexToString
```

```
#Állítsuk elő a classIndexerünket

classIndexer = StringIndexer(inputCol = 'class', outputCol ='indexedclass')
```

```
#Az assemblerünknek kizárólag az oszlopainkat kell átadnunk, mivel előrejelzéseinket az összes oszlop alapján szeretnénk előállítani.

assembler = VectorAssembler(inputCols=['aa_000', 'ab_000', 'ac_000', 'ad_000', 'ae_000', 'af_000', 'ag_000', 'ag_001', 'ag_002', 'ag_003', 'ag_004', 'ag_005', 'ag_006', 'ag_007', 'ag_008', 'ag_009', 'ah_000', 'ai_000', 'aj_000', 'ak_000', 'al_000', 'am_0', 'an_000', 'ao_000', 'ap_000', 'aq_000', 'ar_000', 'as_000', 'at_000', 'au_000', 'av_000', 'ax_000', 'ay_000', 'ay_001', 'ay_002', 'ay_003', 'ay_004', 'ay_005', 'ay_006', 'ay_007', 'ay_008', 'ay_009', 'az_000', 'az_001', 'az_002', 'az_003', 'az_004', 'az_005', 'az_006', 'az_007', 'az_008', 'az_009', 'ba_000', 'ba_001', 'ba_002', 'ba_003', 'ba_004', 'ba_005', 'ba_006', 'ba_007', 'ba_008', 'ba_009', 'bb_000', 'bc_000', 'bd_000', 'be_000', 'bf_000', 'bg_000', 'bh_000', 'bi_000', 'bj_000', 'bk_000', 'bl_000', 'bm_000', 'bn_000', 'bo_000', 'bp_000', 'bq_000', 'br_000', 'bs_000', 'bt_000', 'bu_000', 'bv_000', 'bx_000', 'by_000', 'bz_000', 'ca_000', 'cb_000', 'cc_000', 'cd_000', 'ce_000', 'cf_000', 'cg_000', 'ch_000', 'ci_000', 'cj_000', 'ck_000', 'cl_000', 'cm_000', 'cn_000', 'cn_001', 'cn_002', 'cn_003', 'cn_004', 'cn_005', 'cn_006', 'cn_007', 'cn_008', 'cn_009', 'co_000', 'cp_000', 'cq_000', 'cr_000', 'cs_000', 'cs_001', 'cs_002', 'cs_003', 'cs_004', 'cs_005', 'cs_006', 'cs_007', 'cs_008', 'cs_009', 'ct_000', 'cu_000', 'cv_000', 'cx_000', 'cy_000', 'cz_000', 'da_000', 'db_000', 'dc_000', 'dd_000', 'de_000', 'df_000', 'dg_000', 'dh_000', 'di_000', 'dj_000', 'dk_000', 'dl_000', 'dm_000', 'dn_000', 'do_000', 'dp_000', 'dq_000', 'dr_000', 'ds_000', 'dt_000', 'du_000', 'dv_000', 'dx_000', 'dy_000', 'dz_000', 'ea_000', 'eb_000', 'ec_00', 'ed_000', 'ee_000', 'ee_001', 'ee_002', 'ee_003', 'ee_004', 'ee_005', 'ee_006', 'ee_007', 'ee_008', 'ee_009', 'ef_000', 'eg_000'],outputCol="features")
```

```
#Az algoritmussal csoportosítjuk adatainkat 
from pyspark.ml.classification import RandomForestClassifier

rf = RandomForestClassifier(featuresCol="features", labelCol="indexedclass")
```

```
from pyspark.ml import Pipeline
pipeline = Pipeline(stages=[classIndexer, assembler, rf])
```

```
#Előállítjuk a modellünket az eddigi számításaink alapján
model = pipeline.fit(traindf)
```

```
#Megkapjuk az előrejelzéseinket.
predictions = model.transform(testdf)
```
